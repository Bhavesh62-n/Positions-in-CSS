2.  Difference between absolute and relative positioning in CSS:


---

1. position: relative

Definition: The element is positioned relative to its normal/static position.

Effect on Layout:

It remains in the document flow.

Its original space is preserved (other elements behave as if it's still in its original position).


Positioning:

You can move the element using top, right, bottom, and left.

The shift is from where the element would normally appear.


Use Case: Slight adjustments without affecting the layout of other elements.





---

2. position: absolute

Definition: The element is positioned relative to the nearest positioned ancestor (an ancestor with position: relative, absolute, or fixed).

Effect on Layout:

It is removed from the normal document flow.

No space is reserved for it; it can overlap other elements.


Positioning:

It can be placed anywhere within the containing block using top, right, bottom, left.

If no positioned ancestor exists, it uses the <html> (viewport) as the reference.


Use Case: When you want precise placement or overlapping of elements.

5. z-index Explanation with Code Example

What is z-index?

The z-index property in CSS controls the stacking order of overlapping elements.

Elements with a higher z-index value appear in front of elements with a lower value.

It only works on positioned elements (position: relative, absolute, fixed, or sticky).

```python
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8" />
  <meta name="viewport" content="width=device-width, initial-scale=1.0"/>
  <title>z-index Example</title>
  <style>
    .container {
      position: relative;
      height: 300px;
      background-color: #f0f0f0;
      margin: 30px;
      border: 1px solid #ccc;
    }

    .box1 {
      position: absolute;
      top: 40px;
      left: 40px;
      width: 200px;
      height: 150px;
      background-color: yellow;
      z-index: 1;
      padding: 10px;
      box-shadow: 0 0 5px rgba(0,0,0,0.3);
    }

    .box2 {
      position: absolute;
      top: 70px;
      left: 70px;
      width: 200px;
      height: 150px;
      background-color: red;
      color: white;
      z-index: 2;
      padding: 10px;
      box-shadow: 0 0 5px rgba(0,0,0,0.5);
    }
  </style>
</head>
<body>

  <div class="container">
    <div class="box1">Box 1 (z-index: 1)</div>
    <div class="box2">Box 2 (z-index: 2)</div>
  </div>

</body>
</html>
```

Explanation of Code

Box 1 (yellow) has z-index: 1

Box 2 (red) has z-index: 2

Because Box 2 has a higher z-index, it appears on top of Box 1, even though itâ€™s defined after in the layout.
